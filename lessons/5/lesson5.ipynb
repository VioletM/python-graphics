# Графика в Python. Занятие 5

Цели занятия:
- Добавить отскок мяча от ракеток.
- Сделать возврат мяча в центр при попадании на левую или правую границы поля.

В данном занятии мы сначала будем работать с нашей игрой, а затем решим несколько задач.

## Самостоятельное движение мяча в пределах поля

Добавить отскок мяча от ракеток можно, если вспомнить как мы это сделали в случае с отскоком от рамок поля. Однако в данном случае функцию coords нужно использовать и для мячика, и для ракеток.

**Подумайте** над этой задачей и **реализуйте** отскок от ракеток. Для этого вам понадобится добавить условия на отскок мяча от ракеток в функцию motion. Не забудьте, что для правильного отскока ракетки должны быть не линиями, а прямоугольниками!

## Возврат мяча в центр

Возврат мяча в центр после проигрыша одного из игрока является неотъемлемой частью нашей игры. На следующем занятии мы реализуем подсчет очков, а сейчас нам нужно подготовить для этого некоторый функционал.

**Измените** вашу программу так, чтобы при столкновении с правой или левой сторонами поля мяч возвращался в центр.

## Задачи

1. *Данное задание выполнить для основной версии игры.* Протестируйте игру на предмет багов, если они есть - исправьте (к примеру, если условия недостаточно полно написаны, мячик может попадать туда, где ему нельзя находиться).

2. *Данное задание выполнить отдельно.* На холсте создайте объект - прямоугольник. Реализуйте перемещение объекта при перемещении мыши. Для получения координат мыши используйте event.x и event.y.

3. (*hard*) Для задачи из пункта 2 сделайте перемещение объекта только при зажатой левой клавише мыши.
